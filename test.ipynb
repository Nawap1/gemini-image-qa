```python
import PIL.Image
import os
import google.generativeai as genai
from dotenv import load_dotenv
from IPython.display import Markdown as md

# Load environment variables
load_dotenv()
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
genai.configure(api_key=GEMINI_API_KEY)

# Initialize Gemini model
model = genai.GenerativeModel(model_name="gemini-2.0-flash-exp")
```

```python
# Load test images
flower_image = PIL.Image.open("sample/flowers.jpeg")
display(flower_image)
```

```python
# Test questions
questions = [
    "Describe this image.",
    "How many flowers are there in this image?",
    "What types of flowers are in this image?",
    "What are the dominant colors in this image?",
    "Describe the composition of this image"
]

for question in questions:
    print(f"\nQuestion: {question}")
    response = model.generate_content([question, flower_image])
    display(md(response.text))
    display(md("---"))
```
